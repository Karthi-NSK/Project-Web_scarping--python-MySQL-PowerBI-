```python
import pandas as pd
import mysql.connector
from mysql.connector import errorcode

# Load the cleaned dataset
data = pd.read_csv('Web_Scarping.csv')

# MySQL database connection
config = {
    "user": "root",
    "password": "Vaiverkanni@4",
    "host": "localhost",
    "database": "Web_Scarping",
    "raise_on_warnings": True
}

# Initialize cursor and connection
cursor = None
connection = None

try:
    # Connect to the database
    connection = mysql.connector.connect(**config)
    cursor = connection.cursor()

    # Insert data into the Sports_data table
    for index, row in data.iterrows():
        cursor.execute("""
            INSERT INTO Sports_data (name, year, wins, losses, ot_losses, pct, gf, ga, diff)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        """, (
            row['name'],
            row['year'],
            row['wins'],
            row['losses'],
            row['ot_losses'],
            row['pct'],
            row['gf'],
            row['ga'],
            row['diff']
        ))

    # Commit the transaction
    connection.commit()

except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Something is wrong with your user name or password")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("Database does not exist")
    else:
        print(err)

finally:
    # Check if cursor and connection exist before closing
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```
