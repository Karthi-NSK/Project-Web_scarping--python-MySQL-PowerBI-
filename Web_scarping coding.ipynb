```python

url = "https://www.scrapethissite.com/pages/forms/?page_num=1&per_page=100"
res = requests.get(url)
soup = BeautifulSoup(res.content,"lxml")
trs = soup.find_all("tr",{"class":"team"})
for tr in trs:
    data = {td.get("class")[0]:td.text.strip() for td in tr.find_all("td")}
    print(data)
    
                     
```

```python
page_num = 1
while True:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page_num}&per_page=100"
    res = requests.get(url)
    soup = BeautifulSoup(res.content,"lxml")
    trs = soup.find_all("tr",{"class":"team"})
    if trs:
        for tr in trs:
            data = {td.get("class")[0]:td.text.strip() for td in tr.find_all("td")}
            print(data)
        page_num += 1    

    else:
        break
```

```python
page_num, datas = 1, []
while True:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page_num}&per_page=100"
    res = requests.get(url)
    soup = BeautifulSoup(res.content,"lxml")
    trs = soup.find_all("tr",{"class":"team"})
    if trs:
        for tr in trs:
            data = {td.get("class")[0]:td.text.strip() for td in tr.find_all("td")}
            datas.append(data)
        page_num += 1    

    else:
        break

df = pandas.DataFrame(datas)
df.to_excel("Web.xlsx", index= False)
```
